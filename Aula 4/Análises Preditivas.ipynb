```python
import yfinance as yf
from prophet import Prophet
from prophet.plot import plot_plotly
```

```python
# Colectando dados
ticker = input("Digite o codigo da acao: ") # PETR4.SA
dados = yf.Ticker(ticker).history("2y")

# Tratamento de dados
# Resentado o indice da tabela
treinamento = dados.reset_index()
treinamento = treinamento[["Date", "Close"]]
treinamento["Date"] = treinamento["Date"].dt.date
treinamento["Date"]
treinamento.columns = ["ds", "y"]

# Criando e Treinando o nosso modelo
# Carregando o modelo de dados
modelo = Prophet()

# Treinando o modelo
modelo.fit(treinamento)

# criar o periodo de previsoes
periodo = modelo.make_future_dataframe(90)

previsoes = modelo.predict(periodo)
previsoes

# VisualizaÃ§Ã£o das previsÃµes
plot_plotly(modelo, previsoes)
```
