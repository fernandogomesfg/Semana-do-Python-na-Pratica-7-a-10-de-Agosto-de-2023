## Passo 1: Buscar informações das acções de forma automática

```python
!pip install yfinance
```

```python
import yfinance as yf
```

```python
dados = yf.Ticker("PETR4.SA").history("6mo")
```

```python
fechamento = dados.Close
```

```python
fechamento.plot()
```

## Passo 2: Criar as análises dos dados
- Cotação Maxima
- Cotação Minima 
- Última Cotacao

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
actual = round(fechamento[-1], 2)
print(f"A Cotação Máxima é: R${maxima}")
print(f"A Cotação Mínima é: R${minima}")
print(f"A Cotação Actual é: R${actual}")
```

## Passo 3: Enviar um e-mail para o meu gestor com os resultados das Análises

- Abrir uma nova aba (ctrl + t)
- Digitar o endereço do Gmail, Enter
- Clicar no botão "Escrever"
- Digitar o destinatário
- Digitar o assunto, tab
- Digitar a mensagem do email
- Clicar em enviar

```python
 !pip install pyautogui
```

```python
import pyautogui
import pyperclip
```

```python
ticker = "PETR4.SA"
mensagem = f"""
Prezado Gestor,
Segue abaixo as análises da ação {ticker} dos últimos seis meses:
Cotação máxima: R${round(maxima,2)}
Cotação mínima: R${round(minima,2)}
Cotação actual: R${round(actual,2)}

Atenciosamente,
Fernando Gomes.
"""
```

```python
# pausa de 3 segundos
pyautogui.PAUSE = 4

# Abrir uma nova aba (ctrl + t)
pyautogui.hotkey("ctrl", "t")

# Digitar o endereço do Gmail, Enter
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# Clicar no botão "Escrever"
pyautogui.click(x=223, y=313)

# Digitar o destinatário, tab
pyperclip.copy("fernandogomesbernardo@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Digitar o assunto, tab
pyperclip.copy("Análises Diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Digitar a mensagem do email
pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Clicar em enviar
pyautogui.click(x=1149, y=955)

# fechar a aba do gmail
pyautogui.hotkey("ctrl", "f4")

# Imprimir mensagem de enviado com sucesso
print('E-mail enviado com sucesso!')
```
